```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
link = "../Data/processed/region3_daily.csv"
# data frame
dfRegion3Daily = pd.read_csv(link, index_col='Date', parse_dates=True, dayfirst=True)
dfRegion3Daily.head(3)
```

```python
dfRegion3Daily.plot(subplots=True, figsize=(18.5, 10.5));
```

```python
dfRegion3Daily.plot(figsize=(18.5, 10.5));
```

```python
#set plot size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
#plot
dfRegion3Daily['denguecases'].plot(label="Cluster cases", legend=True)
dfRegion3Daily['Rainfall_mm'].plot(secondary_y=True, label="Rainfall", legend=True)
dfRegion3Daily['PSI'].plot(secondary_y=True, label="PSI", legend=True)
dfRegion3Daily['MaxUVIndex'].plot(secondary_y=True, label="UVI", legend=True)
dfRegion3Daily['Mean_Temperature_C'].plot(secondary_y=True, label="Temperature", legend=True)
```

```python
link = "../Data/processed/region3_month.csv"
# data frame
dfRegion3Monthly = pd.read_csv(link, index_col='month')
dfRegion3Monthly.head(3)
```

```python
dfRegion3Monthly.plot(subplots=True, figsize=(18.5, 10.5));
```
