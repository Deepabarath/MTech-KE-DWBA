```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
link = "../Data/processed/region4_daily.csv"
# data frame
dfRegion4Daily = pd.read_csv(link, index_col='Date', parse_dates=True, dayfirst=True)
dfRegion4Daily.head(3)
```

```python
dfRegion4Daily.plot(subplots=True, figsize=(18.5, 10.5));
```

```python
dfRegion4Daily.plot(figsize=(18.5, 10.5));
```

```python
#set plot size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
#plot
dfRegion4Daily['denguecases'].plot(label="Cluster cases", legend=True)
dfRegion4Daily['Rainfall_mm'].plot(secondary_y=True, label="Rainfall", legend=True)
dfRegion4Daily['PSI'].plot(secondary_y=True, label="PSI", legend=True)
dfRegion4Daily['MaxUVIndex'].plot(secondary_y=True, label="UVI", legend=True)
dfRegion4Daily['Mean_Temperature_C'].plot(secondary_y=True, label="Temperature", legend=True)
```

```python
link = "../Data/processed/region4_month.csv"
# data frame
dfRegion4Monthly = pd.read_csv(link, index_col='month')
dfRegion4Monthly.head(3)
```

```python
dfRegion4Monthly.plot(subplots=True, figsize=(18.5, 10.5));
```
