```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
link = "../Data/processed/region5_daily.csv"
# data frame
dfRegion5Daily = pd.read_csv(link, index_col='Date', parse_dates=True, dayfirst=True)
dfRegion5Daily.head(3)
```

```python
dfRegion5Daily.plot(subplots=True, figsize=(18.5, 10.5));
```

```python
dfRegion5Daily.plot(figsize=(18.5, 10.5));
```

```python
#set plot size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
#plot
dfRegion5Daily['denguecases'].plot(label="Cluster cases", legend=True, kind='bar')
```

```python
#set plot size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
#plot
dfRegion5Daily['denguecases'].plot(label="Cluster cases", legend=True)
dfRegion5Daily['Rainfall_mm'].plot(secondary_y=True, label="Rainfall", legend=True)
dfRegion5Daily['PSI'].plot(secondary_y=True, label="PSI", legend=True)
dfRegion5Daily['MaxUVIndex'].plot(secondary_y=True, label="UVI", legend=True)
dfRegion5Daily['Mean_Temperature_C'].plot(secondary_y=True, label="Temperature", legend=True)
```

```python
link = "../Data/processed/region5_month.csv"
# data frame
dfRegion5Monthly = pd.read_csv(link, index_col='month')
dfRegion5Monthly.head(3)
```
