```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
link = "../Data/processed/region1_daily.csv"
# data frame
dfRegion1Daily = pd.read_csv(link, index_col='Date', parse_dates=True, dayfirst=True)
dfRegion1Daily.head(3)
```

```python
dfRegion1Daily.plot(subplots=True, figsize=(18.5, 10.5));
```

```python
dfRegion1Daily.plot(figsize=(18.5, 10.5));
```

```python
#set plot size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
#plot
dfRegion1Daily['denguecases'].plot(label="Cluster cases", legend=True, kind='bar')
```

```python
#set plot size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
#plot
dfRegion1Daily['denguecases'].plot(label="Cluster cases", legend=True)
dfRegion1Daily['Rainfall_mm'].plot(secondary_y=True, label="Rainfall", legend=True)
dfRegion1Daily['PSI'].plot(secondary_y=True, label="PSI", legend=True)
dfRegion1Daily['MaxUVIndex'].plot(secondary_y=True, label="UVI", legend=True)
dfRegion1Daily['Mean_Temperature_C'].plot(secondary_y=True, label="Temperature", legend=True)
```

```python
link = "../Data/processed/region1_month.csv"
# data frame
dfRegion1Monthly = pd.read_csv(link, index_col='month')
dfRegion1Monthly.head(3)
```
